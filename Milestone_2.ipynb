```python
# âœ… Step 1: Import Libraries
from google.colab import files
import docx

# âœ… Step 2: Upload DOCX file
print("ğŸ“‚ Please upload your DOCX file...")
uploaded = files.upload()

# âœ… Step 3: Read and Display the Text Content
for filename in uploaded.keys():
    try:
        # Load the Word document
        doc = docx.Document(filename)

        # Extract text from all paragraphs
        full_text = [para.text for para in doc.paragraphs]

        # Join paragraphs into a single string
        extracted_text = "\n".join(full_text)

        # Print preview (first 500 characters)
        print(f"\nâœ… --- Extracted from {filename} ---\n")
        print(extracted_text[:500])  # show first 500 chars

    except Exception as e:
        print(f"\nâŒ Error reading {filename}: {e}")
```

```python


!pip install PyPDF2 pdfplumber python-docx

import PyPDF2, pdfplumber, docx

def extract_text_from_txt(path):
    with open(path, "r", encoding="utf-8") as f:
        return f.read()

def extract_text_from_docx(path):
    doc = docx.Document(path)
    return "\n".join([para.text for para in doc.paragraphs])

def extract_text_from_pdf_pypdf2(path):
    text = ""
    with open(path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text += page.extract_text() or ""
    return text

def extract_text_from_pdf_pdfplumber(path):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text
```

```python
from google.colab import files
import os

uploaded = files.upload()

def extract_text_auto(filename):
    ext = os.path.splitext(filename)[-1].lower()
    if ext == ".txt":
        return extract_text_from_txt(filename)
    elif ext == ".docx":
        return extract_text_from_docx(filename)
    elif ext == ".pdf":
        # Try pdfplumber first (better layouts)
        return extract_text_from_pdf_pdfplumber(filename)
    else:
        raise ValueError("Unsupported file type")

for filename in uploaded.keys():
    text = extract_text_auto(filename)
    print(f"\n--- Full Extracted Text from {filename} ---\n")
    print(text[:1000])  # show preview (first 1000 chars)
```

```python
uploaded = files.upload()

combined_text = ""

for filename in uploaded.keys():
    extracted = extract_text_auto(filename)
    combined_text += f"\n\n--- File: {filename} ---\n\n"
    combined_text += extracted

print("\n--- Combined Extracted Text ---\n")
print(combined_text[:2000])
```

```python

output_file = "extracted_text_output.txt"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(combined_text)

print(f" Extracted text saved to {output_file}")
```
