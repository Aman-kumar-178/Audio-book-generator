```python
!pip install PyPDF2 pdfplumber python-docx

import PyPDF2, pdfplumber, docx

def extract_text_from_txt(path):
    with open(path, "r", encoding="utf-8") as f:
        return f.read()

def extract_text_from_docx(path):
    doc = docx.Document(path)
    return "\n".join([para.text for para in doc.paragraphs])

def extract_text_from_pdf_pypdf2(path):
    text = ""
    with open(path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text += page.extract_text() or ""
    return text

def extract_text_from_pdf_pdfplumber(path):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text
```

```python

# Install required packages (run this first)
!sudo apt install tesseract-ocr
!pip install pytesseract pillow pdfplumber

import pytesseract
from PIL import Image
import pdfplumber
import os

def extract_with_ocr(pdf_path):
    """Extract text from scanned PDF using OCR"""
    text = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            print(f"Processing {len(pdf.pages)} pages with OCR...")

            for i, page in enumerate(pdf.pages, 1):
                print(f"  - Analyzing page {i}...")

                # Convert page to image and perform OCR
                img = page.to_image(resolution=300).original
                page_text = pytesseract.image_to_string(img)
                text += f"\n--- Page {i} ---\n{page_text}\n"

        return text
    except Exception as e:
        return f"Error during OCR processing: {str(e)}"

def extract_text_from_pdf_pdfplumber(pdf_path):
    """Extract text from searchable PDF (non-scanned)"""
    text = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            print(f"Extracting text from {len(pdf.pages)} pages...")

            for i, page in enumerate(pdf.pages, 1):
                page_text = page.extract_text()
                if page_text:
                    text += f"\n--- Page {i} ---\n{page_text}\n"
                else:
                    text += f"\n--- Page {i} (No text found, might be scanned) ---\n"

        return text
    except Exception as e:
        return f"Error during PDF extraction: {str(e)}"

def smart_pdf_extraction(pdf_path):
    """Smart extraction that tries regular method first, then OCR fallback"""
    print(f"\n Analyzing: {pdf_path}")
    print("=" * 50)

    # First attempt: Regular text extraction
    print(" Attempting regular text extraction...")
    text = extract_text_from_pdf_pdfplumber(pdf_path)

    # Check if we got meaningful text
    if text and len(text.strip()) > 50:  # If we have substantial text
        print(" Text extracted successfully (searchable PDF)")
        return text, "searchable"
    else:
        # Fallback to OCR
        print(" Little/no text found. Switching to OCR for scanned PDF...")
        ocr_text = extract_with_ocr(pdf_path)
        print(" OCR processing completed")
        return ocr_text, "scanned"

def save_results(text, filename, output_dir="extracted_results"):
    """Save extracted text to a file"""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    output_filename = f"{os.path.splitext(filename)[0]}_extracted.txt"
    output_path = os.path.join(output_dir, output_filename)

    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(text)

    return output_path

# MAIN PROGRAM
def main():
    print(" PDF Text Extractor with OCR Fallback ")
    print("=" * 60)

    # File upload
    uploaded = files.upload()

    if not uploaded:
        print(" No files uploaded. Exiting.")
        return

    results = {}

    for filename in uploaded.keys():
        print(f"\n{'='*60}")
        print(f" PROCESSING: {filename}")
        print(f"{'='*60}")

        # Perform extraction
        extracted_text, pdf_type = smart_pdf_extraction(filename)

        # Save results
        saved_path = save_results(extracted_text, filename)

        # Store results
        results[filename] = {
            'text': extracted_text,
            'type': pdf_type,
            'saved_path': saved_path,
            'char_count': len(extracted_text)
        }

        # Display preview
        print(f"\n EXTRACTION SUMMARY:")
        print(f"   PDF Type: {pdf_type.upper()}")
        print(f"   Characters extracted: {len(extracted_text):,}")
        print(f"   Saved to: {saved_path}")

        print(f"\nTEXT PREVIEW (first 1000 characters):")
        print("-" * 40)
        print(extracted_text[:1000] + "..." if len(extracted_text) > 1000 else extracted_text)
        print("-" * 40)

    # Final summary
    print(f"\n{'='*60}")
    print(" EXTRACTION COMPLETED - SUMMARY")
    print(f"{'='*60}")

    for filename, result in results.items():
        print(f" {filename}:")
        print(f"   Type: {result['type']} | Chars: {result['char_count']:,} | Saved: {result['saved_path']}")

    return results

# Run the program
if __name__ == "__main__":
    extraction_results = main()
```

```python

import re

def clean_text(text):
    # Remove multiple spaces and newlines
    text = re.sub(r'\s+', ' ', text)

    # Remove page numbers (naive: standalone digits)
    text = re.sub(r'\b\d+\b', '', text)

    # Remove weird characters (non-ascii except basic punctuation)
    text = re.sub(r'[^a-zA-Z0-9.,!?;:\'\"()\-\s]', '', text)

    return text.strip()

# Example
raw_text = """
Page 1

This   is     a sample document!!
It has multiple   spaces, 123 numbers, and weird symbols ‚òÖ‚òÖ‚òÖ.

Page 2
More text here...
"""
print("Before Cleaning:\n", raw_text[:300])
print("\nAfter Cleaning:\n", clean_text(raw_text)[:300])
```

```python
import re

def split_into_sentences(text):
    # Basic sentence splitting using punctuation
    sentences = re.split(r'(?<=[.!?]) +', text)
    return [s.strip() for s in sentences if s.strip()]

sample_text = "Hello world! This is session 4. We're splitting sentences? Yes, we are."
sentences = split_into_sentences(sample_text)

print("Original:", sample_text)
print("\nSplit Sentences:")
for s in sentences:
    print("-", s)
```
